```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import requestes

# url="https://hands-on.cloud/wp-content/uploads/2022/01/k-mean-dataset-mall-customer.csv"

dataset2 =pd.read_csv("k-mean-dataset-mall-customer.csv")
```

```python
dataset2 .head(3)
```

```python
# stroing two attributes
# X = dataset2.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']]

X=dataset2[['Annual Income (k$)', 'Spending Score (1-100)']]

# importing the module
import matplotlib.pyplot as plt 

# image size
plt.figure(figsize=(10,5))

# ploting scatered graph
plt.scatter(x= X['Annual Income (k$)'], y=X['Spending Score (1-100)'])


plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Annual Income vs Spending Score')
```

```python
# importing the k-means
from sklearn.cluster import KMeans

# cluster size is set to be 3
km = KMeans(n_clusters = 2, random_state=0)
km.fit(X)
```

```python
# ploting the graph of the clusters
plt.figure(figsize=(8,5))

plt.scatter(x= X.iloc[:, 0], y=X.iloc[:, 1], c= km.labels_)

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)');
plt.show()
```

```python
km.cluster_centers_
```

```python
# Within cluster sum squares
wcss = [] 

# for loop
for i in range(1,15):
    km = KMeans(n_clusters= i, random_state=42)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(7,3))
# ploting the elbow graph
plt.plot(range(1,15), wcss, 'og-')

plt.annotate('Elbow', xy=(5, 50000), xytext=(6, 150000), arrowprops=dict(facecolor='blue', shrink=0.05))
#
plt.xlabel("K")
plt.ylabel("Inertia");
```

```python
# k mean algorithm training
km = KMeans(n_clusters = 5).fit(X)

# # ploting the graph
# plt.figure(figsize=(10,5))

# scatter = plt.scatter(x= X.iloc[:, 0], y=X.iloc[:, 1], c= km.labels_)
scatter = plt.scatter(x= X['Annual Income (k$)'], y=X['Spending Score (1-100)'], c= km.labels_, cmap= "plasma" )

plt.scatter(x=km.cluster_centers_[:,0 ],y= km.cluster_centers_[:,1], c='green', s=500, marker='*', edgecolor='black')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# creating the labels for different clusters
# plt.legend(handles=scatter.legend_elements()[0], labels=[0, 1, 2, 3, 4]);

plt.show()
```
